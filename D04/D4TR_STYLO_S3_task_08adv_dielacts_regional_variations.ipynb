## Task 02:  Analyze the use of dialectal features in a corpus of medieval texts to identify regional variations.

Identify and quantify dialectal markers (e.g., vocabulary, pronunciation, grammar) in the texts to understand the geographic distribution of linguistic features. This code analyzes dialectal variations in medieval texts. It identifies the frequency of specific linguistic markers associated with different regions and visualizes their distribution across texts. The core idea is to use these markers as clues to potentially determine the origin or influence of various dialects in medieval writings.

Hints:
- Consult a reference work on medieval dialects to identify relevant features.
- Use text analysis tools to extract and compare the frequency of dialectal markers across texts.


**Libraries:**

* **pandas:** For data manipulation and analysis.
* **nltk:** For tokenizing text and accessing language data.
* **collections:**
   * `Counter`: To count the occurrences of dialectal features.
* **matplotlib.pyplot:** For plotting the results.

**Code Walkthrough:**

1. **Data and Feature Preparation:**
   * A dictionary `dialect_features` stores sets of words that are characteristic of different regions (north, south, east, west).
   * A DataFrame `df` is created with sample medieval texts and their associated regions. This is placeholder dataâ€”in a real-world scenario, you would use a corpus like REM or REF.

2. **Feature Counting:**
   * `count_dialect_features` function:
     * Tokenizes a given text.
     * Counts the occurrences of words from the `dialect_features` dictionary in the text.
   * The function is applied to each text in the DataFrame to create a new column, `feature_counts`, which stores a dictionary of region-feature counts for each text.

3. **Analysis and Visualization:**
   * A `defaultdict` called `region_feature_counts` is used to aggregate feature counts by region.
   * The code iterates over the DataFrame, updating the counts for each region based on the features found in the texts.
   * The results are printed to the console, showing the frequency of each dialectal marker per region.
   * A bar chart is created to visualize the distribution of dialectal features across the regions.

**Enhancements for Real-World Analysis:**

* **Robust Feature Set:** Expand the `dialect_features` dictionary with a more comprehensive list of region-specific markers derived from linguistic research or existing corpora like the Regional English Morphology Database (REM) or the Regional English Forms (REF) corpus.
* **N-gram Analysis:** Explore the use of n-grams (sequences of n words) as dialectal features, as they can capture more subtle variations than individual words.
* **Statistical Analysis:** Conduct statistical tests (e.g., chi-squared test) to determine if the observed differences in feature frequencies are statistically significant.

```python
import pandas as pd
import nltk
from collections import Counter
import matplotlib.pyplot as plt

# Ensure you have the necessary NLTK data files
nltk.download('punkt')

# Example dialectal features for medieval texts (simplified for demonstration but we have REM/REF corpus! NGRAMS?)
dialect_features = {
    'north': ['ay', 'nay', 'quoth', 'hight'],
    'south': ['ye', 'ne', 'said', 'called'],
    'east': ['you', 'no', 'spoke', 'named'],
    'west': ['thou', 'nae', 'spake', 'clept']
}

# Dummy data representing medieval texts with dialectal features
data = {
    'text': [
        "Ay, I quoth thee, hight thou in the north.",
        "Ye shall ne pass, said the southern lord.",
        "You spoke well, named the eastern knight.",
        "Thou shalt nae pass, spake the western guard.",
        "Nay, I hight thee, quoth the northman again.",
        "Ne shall ye enter, said the southern soldier.",
        "No, you named him rightly, spoke the eastern squire.",
        "Thou art clept, nae more, spake the western elder."
    ],
    'region': [
        'north',
        'south',
        'east',
        'west',
        'north',
        'south',
        'east',
        'west'
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to count dialectal features in a text
def count_dialect_features(text, dialect_features):
    words = nltk.word_tokenize(text.lower())
    feature_counts = {region: 0 for region in dialect_features}
    for word in words:
        for region, features in dialect_features.items():
            if word in features:
                feature_counts[region] += 1
    return feature_counts

# Analyze the texts
df['feature_counts'] = df['text'].apply(lambda x: count_dialect_features(x, dialect_features))

# Summarize the feature counts by region
region_feature_counts = {region: Counter() for region in dialect_features}

for index, row in df.iterrows():
    for region in dialect_features:
        region_feature_counts[row['region']].update(row['feature_counts'])

# Print the results
for region, counts in region_feature_counts.items():
    print(f"Region: {region}")
    for feature, count in counts.items():
        print(f"  {feature}: {count}")

# Plot the results
for region, counts in region_feature_counts.items():
    plt.bar(counts.keys(), counts.values(), label=region)

plt.xlabel('Regions')
plt.ylabel('Frequency of Dialectal Markers')
plt.title('Dialectal Features in Medieval Texts')
plt.legend()
plt.show()
```
