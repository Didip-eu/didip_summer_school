## Task 03: Analyze the vocabulary richness of a medieval poem using type-token ratio.

Calculate the type-token ratio (TTR) of the poem, which measures the diversity of its vocabulary (types = unique words, tokens = total words). The Type-Token Ratio (TTR), a measure of lexical diversity, both overall and as a moving average. Additionally, it analyzes word frequencies, unique words, and word length distributions to gain insights into the poem's linguistic characteristics.

Hints:

- A higher TTR indicates greater vocabulary richness.

**Code Walkthrough:**

2. **Type-Token Ratio (TTR) Analysis:**
   * The `calculate_ttr` function calculates the TTR for a given text.
   * The `moving_average_ttr` function calculates TTR over a sliding window of words.

```python
import re
from collections import Counter
import matplotlib.pyplot as plt

# Some sample medieval poem (Middle English version of "Sumer Is Icumen In")
medieval_poem = """
Sumer is icumen in,
Lhude sing cuccu!
Groweþ sed and bloweþ med
And springþ þe wude nu,
Sing cuccu!

Awe bleteþ after lomb,
Lhouþ after calue cu.
Bulluc sterteþ, bucke uerteþ,
Murie sing cuccu!
Cuccu, cuccu, wel singes þu cuccu;
Ne swik þu nauer nu.

Sing cuccu nu. Sing cuccu.
Sing cuccu. Sing cuccu nu!
"""

# Function to normalize text
def normalize_text(text):
    # Convert to lowercase and remove punctuation
    text = re.sub(r'A REGEX PATTERN IS MUSSING FROM HERE', '', text.lower())
    return text

# Function to calculate Type-Token Ratio (TTR)
def calculate_ttr(text):
    words = "WHAT SHOULD WE DO WITH THE TEXT TO GET WORDS"
    types = len(set(words))
    tokens = len(words)
    return types / tokens if tokens > 0 else 0

# Function to calculate moving average TTR
def moving_average_ttr(text, window_size=50):
    words = text.split()
    ttrs = []
    for i in range(len(words) - window_size + 1):
        window = words[i:i+window_size]
        ttr = calculate_ttr(' '.join(window))
        ttrs.append(ttr)
    return ttrs

# Normalize the poem
normalized_poem = normalize_text(medieval_poem)

# Calculate overall TTR
overall_ttr = calculate_ttr(normalized_poem)

# Calculate moving average TTR
window_size = 20  # Adjust this value as needed
moving_avg_ttr = moving_average_ttr(normalized_poem, window_size)

# Print results
print(f"Original Medieval Poem:\n{medieval_poem}\n")
print(f"Normalized Poem:\n{normalized_poem}\n")
print(f"Overall Type-Token Ratio: {overall_ttr:.4f}")

# Word frequency analysis
word_freq = Counter(normalized_poem.split())
print("\nTop 10 most frequent words:")
for word, count in word_freq.most_common(10):
    print(f"{word}: {count}")

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(range(len(moving_avg_ttr)), moving_avg_ttr)
plt.title(f"Moving Average Type-Token Ratio (Window Size: {window_size})")
plt.xlabel("Word Position")
plt.ylabel("TTR")
plt.grid(True)
plt.show()

# Unique words visualization
unique_words = set(normalized_poem.split())
print(f"\nNumber of unique words: {len(unique_words)}")
print("Unique words:")
print(", ".join(sorted(unique_words)))

# Word length distribution
word_lengths = [len(word) for word in normalized_poem.split()]
plt.figure(figsize=(10, 5))
plt.hist(word_lengths, bins=range(1, max(word_lengths) + 2), align='left', rwidth=0.8)
plt.title("Word Length Distribution")
plt.xlabel("Word Length")
plt.ylabel("Frequency")
plt.xticks(range(1, max(word_lengths) + 1))
plt.grid(True)
plt.show()
```
